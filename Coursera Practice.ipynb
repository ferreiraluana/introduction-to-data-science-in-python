### Creating a function that returns a dataframe with the max/min value, the variation, and the the principal statistical values for the period in analysis

ps.: it uses Statistics Module

```python
def currencyVariation(df):   
    
    currencies = pd.Series(df.columns[1:], name = 'Currency')
    maximum = pd.Series([df[currency].max() for currency in df.columns[1:]], name = 'Max')
    minimum = pd.Series([df[currency].min() for currency in df.columns[1:]], name = 'Min')
    
    variation = pd.Series([i-j for i,j in zip(maximum,minimum)], name = 'Variation')
    sDeviation = pd.Series([statistics.stdev(df[currency]) for currency in df.columns[1:]], name = "Standard Deviation")
    Mean = pd.Series([statistics.mean(df[currency]) for currency in df.columns[1:]], name = "Mean")
    Median = pd.Series([statistics.median(df[currency]) for currency in df.columns[1:]], name = "Median")
    #Mode = pd.Series([statistics.mode(df[currency]) for currency in df.columns[1:]], name = "Mode")
    Variance = pd.Series([statistics.variance(df[currency]) for currency in df.columns[1:]], name = "Variance")

    variation_df = pd.concat([currencies, maximum, minimum, variation, sDeviation, Variance, Mean, Median], axis = 1)
    
    return variation_df

currencyVariation(df)
```
